```python
import numpy as np

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier


with open('traindata.txt', 'r') as file:
    # Read all lines from the file
    data = file.readlines()

for i in range(len(data)):
    parts = data[i].split(',')
    data[i] = [float(part) for part in parts]
    data[i] = np.array(data[i])

with open('trainlabels.txt', 'r') as file:
    labels = [line.strip() for line in file]

for i in range(len(labels)):
    parts = labels[i].split('\n')
    labels[i] = [int(part) for part in parts if part]


labels = [int(item) for sublist in labels for item in sublist] #removing arrayfrom each index
print(labels)





scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)

pca = PCA(n_components=2) 
principal_components = pca.fit_transform(data_standardized)

# print("Principal Components:\n", principal_components)
# print(len(principal_components))


# plt.scatter(principal_components[:, 0], principal_components[:, 1])
# plt.xlabel('Principal Component 1')
# plt.ylabel('Principal Component 2')
# plt.title('PCA Result')
# plt.show()

kmeans = KMeans(n_clusters=20)  # Choose the number of clusters
kmeans.fit(principal_components)
labels = kmeans.labels_
# print(labels)

# plt.scatter(principal_components[:, 0], principal_components[:, 1], c=labels)
# plt.xlabel('Principal Component 1')
# plt.ylabel('Principal Component 2')
# plt.title('K-means Clustering on PCA-transformed Data')
# plt.show()

X_train, X_test, y_train, y_test = train_test_split(principal_components, labels, stratify=labels,random_state=1)
print(y_train)

scaler =StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)

# clf = RandomForestClassifier()
# clf.fit(X_train, y_train)

# # Predict and evaluate
# y_pred = clf.predict(X_test)
# print('Accuracy:', accuracy_score(y_test, y_pred))
MLP=MLPClassifier(hidden_layer_sizes=(180,70,20),max_iter=100,batch_size=16)
MLP.fit(X_train,y_train)
# predictions =MLP.predict(x_values)
prediction=[]
for i in range(len(X_test)):

    prediction.append(MLP.predict([X_test[i]])) 

from sklearn.metrics import classification_report,confusion_matrix
# for i in range(len(y_test)):
#     print(y_test[i])
#     print()
#     print(prediction[i])


print(confusion_matrix(y_test,prediction))
print(classification_report(y_test,prediction))


# print(data)
```
