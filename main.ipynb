```python
# Student Number: 2603828 , 2581366,2540381
import numpy as np
import matplotlib.pyplot as plt

def closest_centroid(centroids, data_point):
    min = 100000000000000000000
    mini = 0
    for i in range(len(centroids)):
        distance = np.linalg.norm(data_point - centroids[i])
        # print(distance)
        if distance < min:
            min = distance
            mini = i
    return mini


def ObjectiveFunction(centroids, clusters):
    sum = 0
    for i in range(len(centroids)):
        for j in range(len(clusters[i])):
            sum += np.linalg.norm(clusters[i][j] - centroids[i])**2 * 0.5
    return sum

k = 4
clust = [[] for i in range(k)]

centers = [(1, -1), (-1, 1), (-1, -1), (1, 1)]

cov = [[1, 0], [0, 1]]

clusters = [np.random.multivariate_normal(center, cov, 20) for center in centers]
```

```python

def Clustering(centroids, Data,k):
    c_data = []
    for m in range(15):
        clust = [[] for i in range(k)]
        for i in range(len(Data)):
            j = closest_centroid(centroids, Data[i])
            clust[j].append(Data[i]) 
            if m==14:
                c_data.append(centroids[j])
        colors = ['red', 'green', 'blue', 'yellow','purple','pink','black','brown']  # Define as many colors as clusters
        for i in range(len(clust)):
            sumn = [0,0,0]
            for j in range(len(clust[i])):
                sumn += clust[i][j]
            if len(clust[i])!=0:
                avg = (1/len(clust[i]))*sumn
                centroids[i] = avg
    return clust,c_data,centroids
```
